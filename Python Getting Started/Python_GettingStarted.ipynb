```
students = []

def get_students_titlecase():
    students_titlecase = [] #Create a list
    for student in students:
        students_titlecase.append(student["name"].title())
    return students_titlecase

def print_students_titlecase():
    students_titlecase = get_students_titlecase()
    print(students_titlecase)

def add_student(name, student_id = 155):
    student = {"name": name, "student_id": student_id}
    students.append(student) #Put dictionary inside a list

def save_file(student):
    try:
        f = open("students.txt", "a") #Here "a" is the access mode means we want to appens to a new or existing file. Selecting "w" is going to overwrite what is already writtten in it
        f.write(student + "\n")
        f.close()
    except Exception:
        print("Could not save file")

def read_file():
    try:
        f = open("students.txt", "r")
        for student in f.readlines():
            add_student(student)
        f.close()
    except Exception:
        print("Could not read file")

read_file()
print_students_titlecase()

student_name = input("Enter student name: ")
student_id = input("Enter student ID: ")
add_student(student_name, student_id)
save_file(student_name)

student_name = input("Enter student name: ")
student_id = input("Enter student ID: ")
add_student(student_name, student_id)

print(students)
print_students_titlecase()
```

Generator Functions - Yield

How do you iterate through a function

```
students = []

def read_file():
    try:
        f = open("students.txt", "r")
        for student in read_students(f):
            add_student(student)
        f.close()
    except Exception:
        print("Could not read file")


def read_students(f):
    for line in f:
        yield line

read_file()
print(students)

```

Inheritance and Polymorphism

```
students =[]

class Student:

    school_name = "Springfield Elementary" #This is a static variable or class attribute that does not change across classes

    def __init__(self, name, student_id = 155): #You ned self for a particular instance of class
        self.name = name #"self.variable" is an instance variable that is available throughout the instance
        self.student_id = student_id
        students.append(self) #Put dictionary inside a list

    def __str__(self):
        return "Student " + self.name

    def get_name_capitalize(self):
        return self.name.capitalize()

    def get_school_name(self):
        return self.school_name

#mark = Student("Mark") #Create object or instantiate a class
#print(mark)
#mark

print(Student.school_name) #Since school_name is a class attribute, it can be directly called from the class


class HighSchoolStudent(Student): #HighSchoolStudent is a derived class whereas Student is a parent class
    school_name = "Springfield High School"

    def get_school_name(self):
        return "This is a High School Student"

    def get_name_capitalize(self):
        original_value = super().get_name_capitalize()
        return original_value + "-HS"

james = HighSchoolStudent("james")
print(james.get_name_capitalize())
```
