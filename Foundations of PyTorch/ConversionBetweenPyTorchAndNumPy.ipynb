Using torch.as_tensor to create a tensor from numpy. 

torch.as_tensor does not create a copy

torch.as_tensor is useful if the data is already a tensor with the same dtype and device then no copy is performed and a new tensor is returned with the same computational graph

```
np_array_one = np.array([4, 8])
np_array_one
```

```
tensor_from_array_one = torch.as_tensor(np_array_one)
tensor_from_array_one
```

```
np_array_one[1] = 5
np_array_one
```

```
tensor_from_array_one
```

Create a copy when making a tensor from numpy array

```
np_array_two = np.array([2, 2])
np_array_two
```

```
tensor_from_array_two = torch.tensor(np_array_two)
tensor_from_array_two
```

Change in the numpy array won't reflect in the tensor

```
np_array_two[1] = 4
np_array_two
```

Here the change won't be reflected

```
tensor_from_array_two
```
